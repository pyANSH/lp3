```python
import pandas as pd

dt = pd.read_csv('https://raw.githubusercontent.com/pyANSH/lp3/main/dataset/emails.csv')
dt.head()
```

```python
dt.isna().sum()
```

```python
import seaborn as sns
sns.heatmap(data=dt.corr(), annot=True)
```

```python
x = dt.drop(columns=['Prediction','Email No.'],axis=1)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier

 sklearn.metrics import accuracy_score
```

```python
y = dt['Prediction']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

### SVM

```python
svc = SVC(C=1.0,kernel='rbf',gamma='auto')         
svc.fit(x_train,y_train)
y_pred2 = svc.predict(x_test)
print("Accuracy Score for SVC : ", accuracy_score(y_pred2,y_test))
```

### KNN

```python
from sklearn.neighbors import KNeighborsClassifier
KNN=KNeighborsClassifier(n_neighbors=7)
KNN.fit(x_train,y_train)
y__pred=KNN.predict(x_test)
print("Accuracy Score for KNN : ", accuracy_score(y__pred,y_test))
```
